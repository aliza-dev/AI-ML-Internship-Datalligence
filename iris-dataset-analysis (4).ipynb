```python
file_path = '../input/iris/Iris.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Id'])
```

```python
df.head()
```

```python
print(f"Count Duplicate rows: {df.duplicated().sum()}")
```

```python
df.drop_duplicates(inplace=True)
print("Duplicate rows are removed")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Species', data=df)
plt.title('Total Flowers in Each Species')
plt.xlabel('Types of Species')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', s=100)  
plt.title('Relationship Between Sepal Length and Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```

```python
print("Relationship Between All Features (Pair Plot):")
sns.pairplot(df, hue='Species')
plt.show()
```
