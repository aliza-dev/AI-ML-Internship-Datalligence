```python
mean= df['Salary'].mean()
df['Salary']= df['Salary'].fillna(mean)
```

```python
median = df['Age'].median()
df['Age']=df['Age'].fillna(median)
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].abs()
```

```python
df['Age']=df['Age'].apply(lambda x: x if 20 < x < 100 else median)
df['Salary'] = df['Salary'].apply(lambda x: round(x, -3) if x >= 500 else 1000)
```

```python
df['Age']=df['Age'].astype(int)
df['Salary']=df['Salary'].astype(int)
```

```python
df['Designation']=df['Designation'].str.title()
```
