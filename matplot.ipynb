```python
plt.scatter(x,y)
plt.show()
```

```python
# Create a scatter plot with Age as red and Fare as green
plt.scatter(x, y, c='red', label='Age')  # Plot Age in red
plt.scatter(x, y, c='green', label='Fare', alpha=0.5)  # Plot Fare in green with some transparency

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

# Add legend to explain the colors
plt.legend(['Age (red)', 'Fare (green)'])

plt.show()
```

```python
# Create a scatter plot with Age as red and Fare as green
plt.scatter(x, y, edgecolor='red', facecolor='green')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.show()
```

```python
# Sample data
x = df['Age']
y = df['Fare']

# Assign colors based on conditions
colors = ['red' if age > 30 else 'green' for age in x]

# Create a scatter plot with specified colors
plt.scatter(x, y, c=colors)

plt.show()
```

```python
# Example: Distribution of passengers by class
class_counts = df['Pclass'].value_counts()

# Create a bar chart
class_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Class')

plt.show()
```

```python
# Example: Number of survivors vs non-survivors
survival_counts = df['Survived'].value_counts()

# Create a bar chart
survival_counts.plot(kind='bar', color=['red', 'green'])
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.title('Survivors vs Non-Survivors')

# Adjust x-ticks to show labels
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])

plt.show()
```

```python
# Count the number of passengers in each class
class_counts = df['Pclass'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen'])
plt.title('Distribution of Passengers by Class')
plt.show()
```

```python
# Count the number of passengers by survival status
survival_counts = df['Survived'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(survival_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Distribution of Passengers by Survival Status')
plt.show()
```
